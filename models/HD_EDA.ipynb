```python
import matplotlib.pyplot as plt
import seaborn as sns
HD_data['age'].hist(bins=20)
plt.show()
HD_data['age'].value_counts().plot(kind='bar')
plt.show()
```

```python
#Distribution plot for a numerical column
sns.histplot(HD_data['age'], kde=True)
plt.show()

#Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(HD_data.corr(), annot=True, fmt='.2f')
plt.show()
```

```python
HD_data.plot.scatter(x='age', y='target')
plt.show()
```

```python
sns.pairplot(HD_data)
plt.show()
```

```python
HD_data.groupby('age')['target'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.show()
```

```python
HD_data.groupby('sex')['target'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.show()
```

```python
for i in HD_data.columns:
    
    HD_data[i].value_counts().plot(kind = "pie", figsize = (6,6), autopct = "%1.1f%%")
    plt.xticks(rotation = 45)
    plt.show()
```

```python
sns.pairplot(HD_data, vars=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal'],hue='target')
```
