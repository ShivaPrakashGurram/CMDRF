```python
LC_data['GENDER']=LC_data['GENDER'].map({'M': 1, 'F': 0})
LC_data['LUNG_CANCER']=LC_data['LUNG_CANCER'].map({'YES': 1, 'NO': 0})
LC_data['SMOKING']=LC_data['SMOKING'].map({2: 1, 1: 0})
LC_data['YELLOW_FINGERS']=LC_data['YELLOW_FINGERS'].map({2: 1, 1: 0})
LC_data['ANXIETY']=LC_data['ANXIETY'].map({2: 1, 1: 0})
LC_data['PEER_PRESSURE']=LC_data['PEER_PRESSURE'].map({2: 1, 1: 0})
LC_data['CHRONIC DISEASE']=LC_data['CHRONIC DISEASE'].map({2: 1, 1: 0})
LC_data['FATIGUE']=LC_data['FATIGUE'].map({2: 1, 1: 0})
LC_data['ALLERGY']=LC_data['ALLERGY'].map({2: 1, 1: 0})
LC_data['WHEEZING']=LC_data['WHEEZING'].map({2: 1, 1: 0})
LC_data['ALCOHOL CONSUMING']=LC_data['ALCOHOL CONSUMING'].map({2: 1, 1: 0})
LC_data['COUGHING']=LC_data['COUGHING'].map({2: 1, 1: 0})
LC_data['SHORTNESS OF BREATH']=LC_data['SHORTNESS OF BREATH'].map({2: 1, 1: 0})
LC_data['SWALLOWING DIFFICULTY']=LC_data['SWALLOWING DIFFICULTY'].map({2: 1, 1: 0})
LC_data['CHEST PAIN']=LC_data['CHEST PAIN'].map({2: 1, 1: 0})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
LC_data['AGE'].hist(bins=20)
plt.show()
LC_data['AGE'].value_counts().plot(kind='bar')
plt.show()
```

```python
#Distribution plot for a numerical column
sns.histplot(LC_data['AGE'], kde=True)
plt.show()

#Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(LC_data.corr(), annot=True, fmt='.2f')
plt.show()
```

```python
LC_data.plot.scatter(x='AGE', y='LUNG_CANCER')
plt.show()
```

```python
sns.pairplot(LC_data)
plt.show()
```

```python
LC_data.groupby('AGE')['LUNG_CANCER'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.show()
```

```python
LC_data.groupby('GENDER')['LUNG_CANCER'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.show()
```

```python
for i in LC_data.columns:
    
    LC_data[i].value_counts().plot(kind = "pie", figsize = (6,6), autopct = "%1.1f%%")
    plt.xticks(rotation = 45)
    plt.show()
```
